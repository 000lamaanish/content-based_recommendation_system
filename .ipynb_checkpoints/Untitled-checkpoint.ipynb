```python
df.drop(columns=['homepage'],inplace=True)
```

```python
df.drop(columns=['tagline'], inplace=True)
```

```python
df['overview'].fillna('No overview available', inplace=True)
```

```python
df['release_date'].fillna("not available",inplace=True)
df['runtime'].fillna("not available", inplace=True)
```

```python
df.isnull().sum()
```

```python
df['budget']=df['budget'].replace(0,np.nan)
```

```python
df['revenue'] = df['revenue'].replace(0, np.nan)
```

```python
df['release_date']
```

```python
df['release_date']=pd.to_datetime(df['release_date'],errors='coerce')
```

```python
import ast
df['genres'] = df['genres'].apply(lambda x: [d['name'] for d in ast.literal_eval(x)] if pd.notnull(x) else [])
```
